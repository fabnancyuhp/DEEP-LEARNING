L’objectif de ce cours est d’être capable d’implémenter des réseaux de neurones avec le langage de programmation Tensorflow.keras. Nous utiliserons différent type de réseaux de neurone :
* ANN (Artificial neural networks)
* CNN : convolutional neural network
* LSTM : Long short-term memory 
* BERT : Bidirectional Encoder Representations from Transformers

Les données utilisées dans ce cours peuvent être de différents types. Nous traitons d’images, de textes, de fichiers CSV ou JSON…

Une connaissance des librairies Sklearn, Pandas et Numpy est un préalable nécessaire pour bien aborder ce cours.


# Environnement de travail
Dans cette UE, il faut faire les codes Python dans des notebooks google colab ou bien des notebooks kaggle. Dans ces notebooks, Tensorflow, Keras ainsi que Pytorch sont préinstallés. On a également accès à de la GPU, ce qui permet d’accélérer les phases d'entraînement des modèles.
